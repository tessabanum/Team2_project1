```python
#Import dependencies
import pandas as pd
```

```python
#Store filepaths(4)
pop_path = "Resources/PopulationEstimates.csv"
unemploy_path = "Resources/Unemployment.csv"
income_path = "Resources/Income.csv"


#Read in csv files(4)
population_df = pd.read_csv(pop_path)
unemployment_df = pd.read_csv(unemploy_path)
income_df = pd.read_csv(income_path)

unemployment_df.head()
```

```python
##Include needed columns ONLY
#Include State, Area_Name, POP_ESTIMATE_2010, POP_ESTIMATE_2011, POP_ESTIMATE_2012, 
    #POP_ESTIMATE_2013, POP_ESTIMATE_2014, POP_ESTIMATE_2015, POP_ESTIMATE_2016, 
    #POP_ESTIMATE_2017, POP_ESTIMATE_2018 in population_df
population_df = population_df[['State', 'Area_Name', 'POP_ESTIMATE_2010', 'POP_ESTIMATE_2011', 
                              'POP_ESTIMATE_2012', 'POP_ESTIMATE_2013', 'POP_ESTIMATE_2014', 
                              'POP_ESTIMATE_2015', 'POP_ESTIMATE_2016', 'POP_ESTIMATE_2017', 
                              'POP_ESTIMATE_2018']]
#Include State, Area_Name, Unemployment_rate_2010, Unemployment_rate_2011, Unemployment_rate_2012, 
    #Unemployment_rate_2013, Unemployment_rate_2014, Unemployment_rate_2015, Unemployment_rate_2016, 
    #Unemployment_rate_2017, Unemployment_rate_2018 in unemployment_df
unemployment_df = unemployment_df[['State', 'Area_name', 'Unemployment_rate_2010', 'Unemployment_rate_2011', 
                                  'Unemployment_rate_2012', 'Unemployment_rate_2013', 'Unemployment_rate_2014', 
                                  'Unemployment_rate_2015', 'Unemployment_rate_2016', 'Unemployment_rate_2017', 
                                  'Unemployment_rate_2018']]
population_df.head()
```

```python
#Rename Area_name to Area_Name
unemployment_df = unemployment_df.rename(columns={'Area_name':'Area_Name'})
unemployment_df.head()
```

```python
#Merge population_df, unemployment_df and income_df on State and Area_Name
pop_employ = pd.merge(population_df, unemployment_df, on=['State', 'Area_Name'])
co_demographics = pd.merge(pop_employ, income_df, on=['State', 'Area_Name'])

co_demographics.head()
```

```python
#write to csv
co_demographics.to_csv("Resources/County Demographics.csv", index=False, header=True)
```
