```python
#Import dependencies
import pandas as pd
```

```python
#Store filepaths(4)
pop_path = "../Resources/PopulationEstimates.csv"
unemploy_path = "../Resources/Unemployment.csv"
income_path = "../Resources/Income.csv"



#Read in csv files(4)
population_df = pd.read_csv(pop_path)
unemployment_df = pd.read_csv(unemploy_path)
income_df = pd.read_csv(income_path)

income_df.head()
```

```python
##Include needed columns ONLY
#Include FIPS, State, Area_Name, POP_ESTIMATE_2010, POP_ESTIMATE_2011, POP_ESTIMATE_2012, 
    #POP_ESTIMATE_2013, POP_ESTIMATE_2014, POP_ESTIMATE_2015, POP_ESTIMATE_2016, 
    #POP_ESTIMATE_2017, POP_ESTIMATE_2018 in population_df
population_df = population_df[['FIPS','Area_Name','State','POP_ESTIMATE_2014']]
#Include FIPS, State, Area_name, Unemployment_rate_2010, Unemployment_rate_2011, Unemployment_rate_2012, 
    #Unemployment_rate_2013, Unemployment_rate_2014, Unemployment_rate_2015, Unemployment_rate_2016, 
    #Unemployment_rate_2017, Unemployment_rate_2018 in unemployment_df
unemployment_df = unemployment_df[['FIPS','Unemployment_rate_2014']]

#Include FIPS, State, Area_Name, HH_income_2014 in income_df
income_df = income_df[['Area_Name','State','HH_income_2014']]

income_df.head()
```

```python
#Rename Area_name to Area_Name
#unemployment_df = unemployment_df.rename(columns={'Area_name':'Area_Name'})

#unemployment_df.head()
```

```python
#Merge population_df, unemployment_df and income_df on State and Area_Name
pop_income = pd.merge(income_df, population_df, on=['Area_Name','State'],how='outer')
co_demographics = pd.merge(pop_income, unemployment_df, on=['FIPS'],how='outer')

co_demographics.head()
```

```python
#rearrange columns
co_demographics = co_demographics[['FIPS','State','Area_Name','POP_ESTIMATE_2014','HH_income_2014','Unemployment_rate_2014']]

co_demographics.head()
```

```python
#write to csv
#unemployment_df.to_csv("Resources/County Unemployment.csv", index=False, header=True)
#population_df.to_csv("Resources/County Population.csv", index=False, header=True)
co_demographics.to_csv("../Resources/County Demographics.csv", index=False, header=True)
```
