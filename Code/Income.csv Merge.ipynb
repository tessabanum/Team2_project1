```python
#Import dependencies
import pandas as pd
```

```python
#Create filepaths (8)
seven_path = "../Resources/est17all.csv"
six_path = "../Resources/est16all.csv"
five_path = "../Resources/est15all.csv"
four_path = "../Resources/est14all.csv"
three_path = "../Resources/est13all.csv"
two_path = "../Resources/est12all.csv"
one_path = "../Resources/est11all.csv"
zero_path = "../Resources/est10all.csv"

#Read csv files
seven_df = pd.read_csv(seven_path)
six_df = pd.read_csv(six_path)
five_df = pd.read_csv(five_path)
four_df = pd.read_csv(four_path)
three_df = pd.read_csv(three_path)
two_df = pd.read_csv(two_path)
one_df = pd.read_csv(one_path)
zero_df = pd.read_csv(zero_path)

seven_df.head()
```

```python
#Include Postal Code, Name, Median Household Income ONLY
seven_df = seven_df[['Postal Code','Name','Median Household Income']]
six_df = six_df[['Postal Code','Name','Median Household Income']]
five_df = five_df[['Postal Code','Name','Median Household Income']]
four_df = four_df[['Postal Code','Name','Median Household Income']]
three_df = three_df[['Postal Code','Name','Median Household Income']]
two_df = two_df[['Postal Code','Name','Median Household Income']]
one_df = one_df[['Postal','Name','Median Household Income']]
zero_df = zero_df[['Postal','Name','Median Household Income']]

zero_df.head()
```

```python
#Rename Median HH Income to include the year
seven_df = seven_df.rename(columns={'Median Household Income':'HH_income_2017'})
six_df = six_df.rename(columns={'Median Household Income':'HH_income_2016'})
five_df = five_df.rename(columns={'Median Household Income':'HH_income_2015'})
four_df = four_df.rename(columns={'Median Household Income':'HH_income_2014'})
three_df = three_df.rename(columns={'Median Household Income':'HH_income_2013'})
two_df = two_df.rename(columns={'Median Household Income':'HH_income_2012'})
one_df = one_df.rename(columns={'Postal':'Postal Code', 'Median Household Income':'HH_income_2011'})
zero_df = zero_df.rename(columns={'Postal':'Postal Code','Median Household Income':'HH_income_2010'})

zero_df.head()
```

```python
#Merge 8 dataframes on Postal Code and Name
zero_one = pd.merge(zero_df, one_df, on=['Postal Code', 'Name'])
zero_two = pd.merge(zero_one, two_df, on=['Postal Code', 'Name'])
zero_three = pd.merge(zero_two, three_df, on=['Postal Code', 'Name'])
zero_four = pd.merge(zero_three, four_df, on=['Postal Code', 'Name'])
zero_five = pd.merge(zero_four, five_df, on=['Postal Code', 'Name'])
zero_six = pd.merge(zero_five, six_df, on=['Postal Code', 'Name'])
hh_income_df = pd.merge(zero_six, seven_df, on=['Postal Code', 'Name'])

hh_income_df.head()
```

```python
#Rename Postal Code and Name columns
hh_income_df = hh_income_df.rename(columns={'Postal Code':'State', 'Name':'Area_Name'})
hh_income_df.head()
```

```python
#Write to csv file
hh_income_df.to_csv("../Resources/Income.csv", index=False, header=True)
```
