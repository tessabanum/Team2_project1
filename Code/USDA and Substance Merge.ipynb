```python
#Import Dependencies
import pandas as pd
```

```python
#Create filepaths
sub_path = '../Resources/substance_data.csv'
census_path = '../Resources/County Demographics.csv'

#Read in csv files
substance_df = pd.read_csv(sub_path)
census_df = pd.read_csv(census_path)
substance_df.head()
```

```python
#Merge 2 datasets on FIPS
cen_sub = pd.merge(census_df, substance_df, on='FIPS', how='outer')
cen_sub.head()
```

```python
#Remove rows where FIPS==NaN
cen_sub = cen_sub[pd.notnull(cen_sub['FIPS'])]
```

```python
#Remove rows where FIPS is duplicated
cen_sub.drop_duplicates('FIPS',inplace=True)
cen_sub.head()
```

```python
#Write to csv file
cen_sub.to_csv('../Resources/Census_Substance2014.csv', index=False, header=True)
```
