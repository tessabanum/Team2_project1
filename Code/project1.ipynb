```python
from scipy.stats import normaltest
from scipy.stats import pearsonr
import pandas as pd
import numpy as np
data_file = "../Data/NewListAll2014-UPDATE01.csv"
data_file_pd = pd.read_csv(data_file)
data_file_pd=pd.DataFrame(data_file_pd)
data_file_pd=data_file_pd[['McDonalds Count','HH_income_2014','Unemployment_rate_2014','Alcohol Mortality Rate','Drug Mortality Rate','Self-harm Mortality Rate','Population']]
#print(data_file_pd.head())
print(data_file_pd.head())
data_cleaned=data_file_pd.dropna(how='any')
print(data_cleaned.head())
McDonalds_cnt=data_cleaned['McDonalds Count']
income=data_cleaned['HH_income_2014']
unemployment=data_cleaned['Unemployment_rate_2014']
population=data_cleaned['Population']
alcohol_death=data_cleaned['Alcohol Mortality Rate']
drug_death=data_cleaned['Drug Mortality Rate']
mental_death=data_cleaned['Self-harm Mortality Rate']
print(data_cleaned.describe())
#income=np.asarray(income)
income=income.tolist()
#data_file_pd_dup=data_file_pd

income=[int(i.replace(',','')) for i in income]
print(income)
```

```python
# normality test-McDonald
stat, p = normaltest(McDonalds_cnt)
print(stat,p)
# interpret
alpha = 0.05
if p > alpha:
    print('Sample looks Gaussian (fail to reject H0)')
else:
    print('Sample does not look Gaussian (reject H0)')

# histogram plot-McDonald
from matplotlib import pyplot

# histogram plot
McDonald_cnt_img=pyplot.hist(McDonalds_cnt)
pyplot.title('McDonalds Count per county')
pyplot.savefig("..\Image\McDonalds Count.png")
pyplot.show()

```

```python
# normality test-Income
stat, p = normaltest(income)
print(stat,p)
# interpret
alpha = 0.05
if p > alpha:
    print('Sample looks Gaussian (fail to reject H0)')
else:
    print('Sample does not look Gaussian (reject H0)')
    
# histogram plot
income_img=pyplot.hist(income,50)
pyplot.title('Income')
pyplot.savefig("..\Image\income.png")
pyplot.show()


#Pearson test
import scipy
CC,P=scipy.stats.pearsonr(McDonalds_cnt,income)
print(f'Correlation of income vs McDonalds counts per county:{CC}, {P}')
```
