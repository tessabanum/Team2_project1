```python
#Dependencies
import pandas as pd
```

```python
#load file 
substance_abuse = "Resources/Substance_Mortality_2014.xlsx"
```

```python
#read multiple into a dataframe
df = pd.read_excel(substance_abuse)

df.head()
```

```python
#create a dataframe to read the first sheet
df_alcohol_use = pd.read_excel(substance_abuse, sheet_name=0, index_col=0)
df_alcohol_use.head()
```

```python
#create a dataframe to read the second sheet
df_drug_use = pd.read_excel(substance_abuse, sheet_name=1, index_col=0)
df_drug_use.head()
```

```python
#create a dataframe to read the third sheet
df_self_harm = pd.read_excel(substance_abuse, sheet_name=2, index_col=0)
df_self_harm.head()
```

```python
#create a datafrane to read the fourth sheet
df_int_violence = pd.read_excel(substance_abuse, sheet_name=3, index_col=0)
df_int_violence.head()
```

```python
df_substance_data = pd.merge(df_alcohol_use, df_drug_use, how="outer", on=["Location", "FIPS", "FIPS"])

df_substance_data.head()
```

```python
#create a dataframe to merge self-harm and Int. viole
df_substance_data_2 = pd.merge(df_self_harm, df_int_violence, how="outer", on=["Location", "FIPS", "FIPS"])

df_substance_data_2.head()
```

```python
#merge the 2 data frame to get 1 result. 
df_substance_data_3 = pd.merge(df_substance_data, df_substance_data_2, how="outer", on=["Location", "FIPS", "FIPS"])

df_substance_data_3
                               
```

```python
#create a dataframe to merge self-harm and Int. viole
df_substance_data_3.to_csv("Resources/substance_data.csv", index=False, header=True)
```
