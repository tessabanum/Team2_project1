```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#file to load
new_list_data = "Resources/NewListAll2014.csv"
```

```python
#read the file 
df_all_data = pd.read_csv(new_list_data)

df_all_data.head()
```

```python
#delete unwanted columns
del df_all_data["Unnamed: 0"]
del df_all_data["Unnamed: 0.1"]

#display result
df_all_data.head()
```

```python
#create a variable for each columns
med_income = df_all_data["HH_income_2014"]
unemploy_rate = df_all_data["Unemployment_rate_2014"]
mcDoc_count = df_all_data["McDonalds Count"]

#create a scatter plot 
bubble_5 = plt.scatter(med_income, unemploy_rate, s = mcDoc_count, color=["red"], edgecolor="black", alpha=0.75, marker="o")
plt.grid(True)
```
